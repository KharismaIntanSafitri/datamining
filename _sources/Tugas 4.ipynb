K-means merupakan salah satu algoritma yang bersifat unsupervised learning. K-Means memiliki fungsi untuk mengelompokkan data kedalam data cluster. Algoritma ini dapat menerima data tanpa ada label kategori. K-Means Clustering Algoritma juga merupakan metode non-hierarchy. 

kali ini membangun model K-Means mengkelompokkan dan memberikan label pada data bunga iris berdasarkan kategori ukuran sepal dan petal. Dataset Iris tersedia di scikit-learn dan kita dapat menggunakannya untuk membangun K-Means kita.

```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
#the imported dataset does not have the required column names so lets add it
colnames = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']
irisdata = pd.read_csv(url, names=colnames)
```

```python
irisdata.head()
```

```python
irisdata['Class'] = pd.Categorical(irisdata["Class"])
irisdata["Class"] = irisdata["Class"].cat.codes
```

```python
X = irisdata.values[:, 0:4]
y = irisdata.values[:, 4]
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
y_true = le.fit_transform(y)
y_true
```

```python
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters=3)
# Fitting the input data
kmeans = kmeans.fit(X)
# Getting the cluster labels
labels = kmeans.predict(X)
# Centroid values
centroids = kmeans.cluster_centers_
```

```python
from sklearn.metrics import classification_report

target_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

print(classification_report(irisdata['Class'],kmeans.labels_,target_names=target_names))
```
