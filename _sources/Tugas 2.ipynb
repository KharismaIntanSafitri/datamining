## **1. Equal Width Intervals**

Equal-width intervals adalah discretization yang membagi data numerik menjadi beberapa kelompok dengan lebar kelompok yang kurang lebih sama besar.

```python
c_panjang_kelopak = dataset[["sepal_length"]]
c_lebar_kelopak = dataset[["sepal_width"]]
c_panjang_daunkelopak = dataset[["petal_length"]]
c_lebar_daunkelopak = dataset[["petal_width"]]
```

Menggunakan fungsi **cut** untuk mencari Equal-Width-Intervals

```python
c_panjang_kelopak["Kategori"] = pd.cut(dataset["sepal_length"],3, labels = ['A','B','C'])
c_lebar_kelopak["Kategori"] = pd.cut(dataset["sepal_width"],3, labels = ['A', 'B','C'])
c_panjang_daunkelopak['Kategori'] = pd.cut(dataset["petal_length"],3, labels = ['A', 'B','C'])
c_lebar_daunkelopak['Kategori'] = pd.cut(dataset["petal_width"],3, labels = ['A', 'B','C'])

equalWidth = pd.DataFrame([])
equalWidth["sepal_length"] = dataset["sepal_length"]
equalWidth['sepal_length_category'] = c_panjang_kelopak["Kategori"]

equalWidth["sepal_width"] = dataset["sepal_width"]
equalWidth["sepal_width_category"] = c_lebar_kelopak["Kategori"]

equalWidth["petal_length"] = dataset["petal_length"]
equalWidth["petal_length_category"] = c_panjang_daunkelopak["Kategori"]

equalWidth["petal_width"] = dataset["petal_width"]
equalWidth["petal_width_category"] = c_lebar_daunkelopak["Kategori"]

equalWidth["species"] = dataset[["species"]]

equalWidth
```

## **3. Entropy**

Entropi adalah nilai informasi yang menyatakan ukuran ketidakpastian(impurity) dari attribut dari suatu kumpulan obyek data dalam satuan bit

dalam tugas kali ini kita akan menggunakan data panjang kelopak (sepal length) sebagai sempel perhitungan

```python
uji = dataset[["sepal_length"]]
uji.describe()
```

membuat category random untuk semua data

```python
np.random.seed(0)
uji["category"] = np.where(np.random.choice(2, uji.shape[0]) < 1, "A", "B")
```

```python
print('Pengelompokan data uji dengan kategori random')
uji
```

membuat fungsi getOverCategory yang digunakan untuk menghitung data keseluruhan yang nantinya digunakan untuk menghitung entropy

```python
def getOverCategory(col):
  group = uji.loc[:, :].groupby("category").count()
  a = group.loc["A", col]
  b = group.loc["B", col]
  return (a, b, a+b)
```

fungsi splitter digunakan untuk membuat split antara value yang telah ditentukan lalu mengembalikan data yang telah dipisahkan

```python
def splitter(value:float, col:str)->tuple:
  # get data less and greater from value
  less = uji[uji[col] <= value]
  greater = uji[uji[col] > value]

  # calculate into category for each data
  less_group = less.loc[:, :].groupby("category").count()
  greater_group = greater.loc[:, :].groupby("category").count()

  # get value based on category
  less_category_A = less_group.loc["A", col] 
  less_category_B = less_group.loc["B", col] 
  greater_category_A = greater_group.loc["A", col] 
  greater_category_B = greater_group.loc["B", col] 

  return (
      [less_category_A, less_category_B, less_category_A + less_category_B],
      [greater_category_A, greater_category_B, greater_category_A + greater_category_B]
  )
```

Membuat fungsi entropy untuk mencari nilai entropy   

Rumus Mencari Entropy :

$$Entropy (D_{1}) = - \sum_{i=1}^{m} pi \log_{2} pi$$ 

```python
def entropy(d):
  r1 = (d[0] / d[2]) * np.log2(d[0] / d[2]) 
  r2 = (d[1] / d[2]) * np.log2(d[1] / d[2]) 
  return np.sum([r1, r2]) * -1
```

Membuat fungsi info

```python
def info(d):
  r1 = (d[0][2] / uji.shape[0]) * entropy(d[0])
  r2 = (d[1][2] / uji.shape[0]) * entropy(d[1])
  return r1 + r2
```

Mmebuat fungsi gain untuk menghitung selisih antara entropy awal dengan yang baru.

Rumus Mencari Gain :

$$Gain(E_{new}) = (E_{initial}) - (E_{new})$$

```python
def gain(Einitial, Enew):
  return Einitial - Enew
```

Membuat DInitial

```python
D = getOverCategory("sepal_length")
entropy_d = entropy(D)
print("D = ",D)
print("Entropy D = ", entropy_d)
```

Melakukan beberapa tes split untuk mencari hasil dan informasi yang terbaik

Tes Pertama : Split 1:4.4

```python
split1  = splitter(4.4, "sepal_length")
info_split1 = info(split1)

print('Gain = ', gain(entropy_d, info_split1))
```

Tes Kedua : Split 2:5.5

```python
split2  = splitter(5.5, "sepal_length")
info_split2 = info(split2)

print('Gain = ', gain(entropy_d, info_split2))
```

Tes Ketiga : Split 3:7.0

```python
split3  = splitter(7.0, "sepal_length")
info_split3 = info(split3)

print('Gain = ', gain(entropy_d, info_split3))
```

Dari seluruh hasil percobaan tes split yang telah dilakukan, maka diperoleh hasil split terbaik adalah split 3 yang mwemberikan keuntungan informasi sebesar 0.0005490214732508658, karena hasil tes split 3 memiliki nilai split yang terendah.
