# **1. Membaca data dari url**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://vincentarelbundock.github.io/Rdatasets/csv/AER/GoldSilver.csv"
dataset = pd.read_csv(url)
```

```python
dataset.info( )
```

```python
data_gold = dataset['gold']
data_gold
```

# **1.2 Data Silver**

```python
data_silver = dataset['silver']
data_silver
```
